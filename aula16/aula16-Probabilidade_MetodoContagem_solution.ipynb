### Biblioteca itertools

A biblioteca itertools te oferece funções que te permite gerar combinações distintas entre elementos de dois conjuntos. Nesta aula, utilizaremos a função itertools.product para gerar tanto resultados de um produto cartesiano quanto potência cartesiana.

```python
import itertools
```

```python
# produto cartesiano
A = { 0, 1}
B = {"a", "b"}

for i in itertools.product(A, B): 
    print(i)
```

```python
# potência cartesiana
A = { 0, 1}

for i in itertools.product(A, repeat=3): 
    print(i)
```
