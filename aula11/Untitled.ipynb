```python
import pandas as pd
```

```python
data = pd.read_csv("japan_birth.tab", sep="\t")
data
```

```python
years = data["year"].drop_duplicates()
years
```

```python
import matplotlib.pyplot as plt
```

```python
for year in years:
    data2 = data[data["year"]==year]
    plt.plot(data2["mouth"],data2["birth"], label=year)

plt.legend(loc=4)
plt.show()
```
